```python
import matplotlib.pyplot as plt
```

```python
import tensorflow.keras as tfk
```

```python
from datasets import load_dataset
import vae
```

```python
train_dataset, val_dataset, test_dataset, dataset_info = vae.Dataset("GNILC", 8)
```

```python
ae = vae.AE(latent_dimension=256)
ae.compile(optimizer=tfk.optimizers.Adam(learning_rate=1e-2, beta_1=0.99, clipnorm=1))
ae.fit(train_dataset, epochs=10, validation_data=val_dataset)
arr = next(iter(test_dataset))
prediction = ae.predict(arr)
fig, ax = plt.subplots(2, 1)
ax[0].imshow(arr[0,:,:,0])
ax[1].imshow(prediction[0,:,:,0])
```
